```python
import pandas as pd

df1 = pd.read_csv("2019.06.05_M3_LTAR001-164.csv")
df = pd.read_csv('2019.06.05_M3_LTAR001-164.csv')


x = 0 #Counter
    
#Getting rid of useless columns

df = df.drop("Type", axis=1)
df = df.drop("Flags", axis=1)
df = df.drop("Soln Conc", axis=1)
df = df.drop("Units", axis=1)
df = df.drop("Units.1", axis=1)
df = df.drop("Int", axis=1)

#Loop for gathering the measurements

for y in df['Element']:
#Al
    if y == 'Al MultiCal' or y == 'Al 394.401':
            AlValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Al MultiCal'] = float(AlValue)
#Ca    
    elif y == 'Ca MultiCal'or y =='Ca 422.673' or y == 'Ca Multical':
            CaValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Ca MultiCal'] = float(CaValue)
#Cu    
    elif y == 'Cu MultiCal' or y == 'Cu 324.754' or y == 'Cu Multical':
            CuValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Cu MultiCal'] = float(CuValue)
#K    
    elif y == 'K MultiCal':
            KValue = df.loc[x,"Corr Con"]
            df.loc[x, 'K MultiCal'] = float(KValue)
#Mg    
    elif y == 'Mg  MultiCal' or y == 'Mg 280.270' or y == 'Mg Multical':
            MgValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Mg MultiCal'] = float(MgValue)
#Mn    
    elif y == 'Mn MultiCal' or y == 'Mn 259.372':
            MnValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Mn MultiCal'] = float(MnValue)
#P    
    elif y == 'P MultiCal' or y == 'P multiCal' or y == 'P multical' or y == 'P Multical':
            PValue = df.loc[x,"Corr Con"]
            df.loc[x, 'P MultiCal'] = float(PValue)
#S    
    elif y == 'S MultiCal' or y == 'S 181.972':
        if y == float or y == int:
            SValue = df.loc[x,"Corr Con"]
            df.loc[x, 'S MultiCal'] = float(SValue)
        else:
            SValue = df.loc[x,"Corr Con"]
            df.loc[x, 'S MultiCal'] = SValue
#Zn  
    elif y == 'Zn MultiCal' or y == 'Zn 206.200':
            ZnValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Zn MultiCal'] = float(ZnValue)
#Fe            
    elif y == 'Fe MultiCal' or y == 'Fe 238.204':
            ZnValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Fe MultiCal'] = float(ZnValue)
#Mo            
    elif y == 'Mo 204.598':
            MoValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Mo MultiCal'] = float(MoValue)
#Ni            
    elif y == 'Ni 230.299':
            NiValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Ni MultiCal'] = float(NiValue)
#Ca
    elif y == 'Na MultiCal':
            NaValue = df.loc[x,"Corr Con"]
            df.loc[x, 'Na MultiCal'] = float(NaValue)
            
    x += 1 #Loop Counter
        
#Dropping the columns element and corr con

df = df.drop("Element", axis = 1)
df = df.drop("Corr Con", axis = 1)

#Splitting and bringing back together the dataframe
group_df = df.groupby('Solution Label')

end_df = group_df.max()

end_df
#end_df.to_excel("2019.06.06_AR_D36 079-110 FIXED.xlsx")
```
